```python
def binning(x,limit1,limit2):
  result="High"
  if x<limit1:
    result="Low"
  elif (x>limit1) & (x<limit2):
    result="Medium"
  return(result)
```

```python
cc_usage["Credit_Limit_segment2"]=cc_usage["Credit_Limit"].apply(lambda x:binning(x,3000,6000))
```
